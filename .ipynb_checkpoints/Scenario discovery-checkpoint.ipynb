# Scenario discovery

In this notebook different, manually set policies are tested an can be compared.
Policies need to be set in the "Possible policies.xlsx" file. The script runs only the new policies added policies when told it so.

First the imports:

```python
from __future__ import (unicode_literals, print_function, absolute_import,
                        division)

%matplotlib inline
from ipywidgets import interact

from ema_workbench import (Model,MultiprocessingEvaluator, SequentialEvaluator, Policy,
                           Scenario)

from ema_workbench.em_framework.evaluators import perform_experiments
from ema_workbench.em_framework.samplers import sample_uncertainties
from ema_workbench.util import ema_logging
from problem_formulation import get_model_for_problem_formulation
from ema_workbench.analysis import prim

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
```

Description:

*run_Scenarios*: if True, then the policies that are not in the previous saved data yet get runned using the multiprocessing evaluator. If False, loads experiments and outcomes from the results folder.
*n_Scenarios*: Number of experiments each alternative should go through
*outcome_option*: The problem statement provides different option of what shall be the level of aggregation of the outcomes.

In the following code, the directory to the "Possible policies.xlsx", experiment data and outcome data can be set.

```python
run_Scenarios=False
n_Scenarios = 1000
outcome_option = 3

df_pol = pd.read_excel(r'Possible policies.xlsx',index_col=0)
try:
    experiments = pd.read_csv('./results/exp_unc_different_pol_{}s_{}.csv'.format(n_Scenarios,outcome_option),index_col=0)
    outcomes = pd.read_csv('./results/out_unc_different_pol_{}s_{}.csv'.format(n_Scenarios,outcome_option),index_col=0)
except:
    experiments = pd.DataFrame(columns=['A.0_ID flood wave shape','A.1_Bmax','A.1_Brate','A.1_pfail','A.2_Bmax','A.2_Brate','A.2_pfail','A.3_Bmax','A.3_Brate','A.3_pfail','A.4_Bmax','A.4_Brate','A.4_pfail','A.5_Bmax','A.5_Brate','A.5_pfail','discount rate 0','discount rate 1','discount rate 2','0_RfR 0','0_RfR 1','0_RfR 2','1_RfR 0','1_RfR 1','1_RfR 2','2_RfR 0','2_RfR 1','2_RfR 2','3_RfR 0','3_RfR 1','3_RfR 2','4_RfR 0','4_RfR 1','4_RfR 2','EWS_DaysToThreat','A.1_DikeIncrease 0','A.1_DikeIncrease 1','A.1_DikeIncrease 2','A.2_DikeIncrease 0','A.2_DikeIncrease 1','A.2_DikeIncrease 2','A.3_DikeIncrease 0','A.3_DikeIncrease 1','A.3_DikeIncrease 2','A.4_DikeIncrease 0','A.4_DikeIncrease 1','A.4_DikeIncrease 2','A.5_DikeIncrease 0','A.5_DikeIncrease 1','A.5_DikeIncrease 2','scenario','policy','model'])
    outcomes = pd.DataFrame()
    print('No saved data found')
    
```

```python
ema_logging.log_to_stderr(ema_logging.INFO)
dike_model, planning_steps = get_model_for_problem_formulation(outcome_option)

if __name__ == '__main__':
    
    uncalc_pol = list(set(df_pol.columns) - set(experiments['policy'].unique()))
    policy_list=[]
    
    for column in uncalc_pol:
        policy_list.app
        end(Policy(column, **df_pol[column].to_dict()))
    
    #In case it was asked for, run the experiments
    if run_Scenarios:
        #Prints start time for performance measurements
        print('Started evaluating at:',datetime.now())
        
        with MultiprocessingEvaluator(dike_model) as evaluator:
            results = evaluator.perform_experiments(scenarios=n_Scenarios, 
                                                    policies=policy_list,
                                                    reporting_frequency=100)
        
        #Of course, a stopping time is also needed
        print('Stopped evaluating at:',datetime.now())
        
        #experiments and outcomes are stored in an csv file
        experiments_new, outcomes_new = results
        outcomes_new = pd.DataFrame(outcomes_new)
        
        experiments_to_save= pd.concat([experiments,experiments_new])
        outcomes_to_save= pd.concat([outcomes,outcomes_new])
        
        experiments_to_save.to_csv('./results/exp_unc_different_pol_{}s_{}.csv'.format(n_Scenarios,outcome_option))
        outcomes_to_save.to_csv('./results/out_unc_different_pol_{}s_{}.csv'.format(n_Scenarios,outcome_option))
        
        outcomes = outcomes_to_save
        experiments = experiments_to_save
```

## Analysis tools for outcome option = 3

this outcome option shows:
- Total cost
- Expected Number of Deaths
for each node and
- Expected Evacuation Costs
globally.

### Total cost

Note: *Evacuation Cost* appear to be not plotted, but infact they are so small compared to the other costs. The y-axis can be adjusted but into a log-mode with the *log* variable set to True.

```python
#prep for Total cost
log = False


df_tmp0 = outcomes.filter(regex='Total Costs')
df_tmp0.columns = list(x.split(' ')[0] for x in df_tmp0.columns)
df_tmp0.loc[:,'Total Cost'] = df_tmp0.filter(regex='A.').sum(axis=1) + outcomes.loc[:,'Expected Evacuation Costs']
df_tmp0.loc[:,'Policy'] = experiments.loc[:,'policy']
df_tmp0.loc[:,'Evacuation Costs'] = outcomes.loc[:,'Expected Evacuation Costs']

df_tmp1 = pd.melt(df_tmp0,
                  value_vars=['A.1', 'A.2', 'A.3', 'A.4', 'A.5', 'RfR', 'Total Cost','Evacuation Costs'],
                  id_vars=['Policy'])

df_tmp1 = df_tmp1.sort_index(axis=1)
sorted_cols = sorted(df_tmp1['Policy'].unique())

fig, ax = plt.subplots(1, figsize=(16,8))
sns.barplot(x='variable',
            y='value',
            data=df_tmp1,
            hue='Policy',
            hue_order=sorted_cols,
            ax=ax,
            palette='Paired'
           )
if log:
    ax.set(yscale='log')
ax.set_xlabel('')
ax.set_ylabel('Costs / €')
```

### Expected Number of Deaths

```python
df_tmp0 = outcomes.filter(regex='Expected Number of Deaths')
df_tmp0.columns = list(x.split('_')[0] for x in df_tmp0.columns)
df_tmp0.loc[:,'Total Deaths'] = df_tmp0.sum(axis=1)
df_tmp0.loc[:,'Policy'] = experiments.loc[:,'policy']


df_tmp1 = pd.melt(df_tmp0,
                  value_vars=['A.1', 'A.2', 'A.3', 'A.4', 'A.5','Total Deaths'],
                  id_vars=['Policy'])

df_tmp1 = df_tmp1.sort_index(axis=1)
sorted_cols = sorted(df_tmp1['Policy'].unique())

fig, ax = plt.subplots(1, figsize=(16,8))
sns.barplot(x='variable',
            y='value',
            data=df_tmp1,
            hue='Policy',
            hue_order=sorted_cols,
            palette='Paired',
            ax=ax)

ax.set_xlabel('')
ax.set_ylabel('Deaths per year')
```
